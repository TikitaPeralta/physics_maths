# Gaussian/Normal Formula:

$$P(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{\frac{-(x-\mu)^2}{2\sigma^2}}$$

### Calculating the Mean:

$$\mu = \frac{1}{N} \sum_{i = 1}^{N} x_{i}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
  
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-4, 4, 0.01)
  
# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
plt.xlabel('x')
plt.ylabel('Probability')
plt.vlines(0, 0.04, 0.173, color = 'grey', linestyle='dotted')
plt.annotate('1 Standard Deviation', xy=(0,0.1), xytext=(2.5,0.1), arrowprops=dict(arrowstyle='<->'))
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

$$\sigma(x) = \sqrt{\frac{\sum_{i=1}^{N}(x_{i}-\mu)^2}{N-1}}$$
